```python
import pandas as pd
```

## COUNTRY

```python
country = pd.read_csv('csv/country31may2015.csv',
                     dtype={'CIKNUMBER': 'S',
                           'IRSNUMBER': 'S',
                           'TAXHAVEN': 'S'} )
```

CIKNUMBER - chamber of commerce nr <br>
IRSNUMBER - tax number <br>
NAME - company name <br>
DATADATE - date of establishment <br>
FILEDATE - file date <br>
INCSTATE - state of incorporation <br>
HQSTATE - state where headquarters is <br>
ISO3 - subsidiary ISO3 <br>
COUNTRY - subsidiary company country <br>
TAXHAVEN - subsidiary company a tax haven? <br>
COUNTRYCOUNT - amount of branches in the country <br>
TOTALCOUNT - total amount of subsidiary branches <br>
NCOUNTRIES - total amount of countries with subsidiaries <br>


### descriptives

TOTAL 447,956 records <br>
most subsidiaries belong to AON CORP (1572) <br>
most subsidiaries are settled in Canada (24884) <br>
median amount of branches per country is 2, max 650 <br>
median amount of subsidiaries per company is 48, max 2085 <br>
median amount of countries a company has subsidiaries in is 12, max 106 <br>
all distributions are highly skewed, meaning that there are a few companies with lots of subsidiaries <br>
there are 90,727 subsidiaries located in tax havens

```python
country.describe(include="all").fillna(value='')
```

```python
country.groupby(['TAXHAVEN'])['TAXHAVEN'].count()
```

### most used tax havens

```python
country[country['TAXHAVEN'] == '1.0'].groupby(['COUNTRY']).size().sort_values(ascending=False)
```
